```python
def gaussian_fun(x,s,m):
    ans = ( 1./(s * np.sqrt(2*np.pi))) * np.exp( (-0.5) * ((x-m)/s)**2 )
    return ans
```

```python
import numpy as np
import random

x = []

for i in range(10000):
    x.append(random.random() * 1000)
    
x.sort()
x = np.array(x)

m = float(x.mean())
s = float(x.std())
y =  gaussian_fun(x,s,m)
```

```python
def braking_distance(velocity, coeff):
    return (1/2)*( velocity**2 / (coeff * 9.8))
```
