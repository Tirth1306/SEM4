# Practical 7

Scan an integer n from the user. Scan n sentences (no special character,
punctuation, all lower case letters). Each sentence’s sentiment is either
positive or negative. Scan 2 different words y and z from the user. y and z
should be words from the set of distinct words constructed from n
sentences. Write a program to estimate (i) the probability of y being
present in the positive sentences (ii) the joint probability of y and z being
present in positive sentences assuming that the presence of y and z in any
sentence are independent events. What would be this probability if these
events are not independent? 

```python
from collections import Counter
```

```python
n = int(input("Enter the number of sentences: "))
```

```python
sen = []
positive = Counter()
negative = Counter()
for i in range(n):
    temp = input(f"Enter the sentence no. {i+1}: ").lower()
    temp = temp.replace('.'," ")
    temp = temp.replace(','," ")
    temp = temp.replace('!'," ")
    temp = temp.replace('?'," ")
    sen.append(temp.strip())
    
    while True:
        sent = input(f"Enter the sentiment of the above sentence (either positive or negative): ")
        sent = sent.strip().lower()

        if(sent == 'positive'):
            positive.update(temp.split(" "))
            break
        elif(sent == 'negative'):
            negative.update(temp.split(" "))
            break
        else:
            print("Invalid input..Try Again")
            
    print("\n\n")
```

```python
y = input("Enter the word y: ").lower()

while ( (y not in positive.keys()) and (y not in negative.keys()) ):
    print("Word data not available")
    y = input("Enter the word y: ").lower()
    
z = input("Enter the word z: ").lower()

while ( (z not in positive.keys()) and (z not in negative.keys()) ):
    print("Word data not available")
    z = input("Enter the word z: ").lower()
```

(i) the probability of y being present in the positive sentences

```python
print("The probability of y being present in the positive sentences")
print(f"Probability is { positive[y]/ (positive[y] + negative[y])}")
```

(ii) the joint probability of y and z being present in positive sentences assuming that the presence of y and z in any sentence are independent events.

```python
print("the joint probability of y and z being present in positive sentences assuming that the presence of y and z in any sentence are independent events")
print(f"Probability is { (positive[y]/ (positive[y] + negative[y])) * (positive[z]/ (positive[z] + negative[z]))}")
```

# Using Real Data

```python
import pandas as pd
```

```python
f = pd.read_csv('amazon_cells_labelled.txt',delimiter = '\t')
f2 = pd.read_csv('yelp_labelled.txt',delimiter = '\t')
f3 = pd.read_csv('imdb_labelled.txt',delimiter = '\t')
```

```python
f = f.append(f2)
f = f.append(f3)
f= f.reset_index()
del f['index']
f['sentiment'] = f['bool'].map(lambda x: 'positive' if x == 1 else 'negative')
del f['bool']
```

```python
pos = Counter()
neg = Counter()
```

```python
for i in range(len(f)):
    temp = f['Sentence'][i].lower()
    temp = temp.replace('.'," ")
    temp = temp.replace(','," ")
    temp = temp.replace('!'," ")
    temp = temp.replace('?'," ")
    temp = temp.replace('('," ")
    temp = temp.replace(')'," ")
    
    if(f['sentiment'][i] == 'negative'):
        neg.update(temp.split(" "))
    else:
        pos.update(temp.split(" "))
```

```python
print("Positive: ")
print(pos.keys())

print("Negative: ")
print(neg.keys())
```

```python
y = input("Enter the word y: ").lower()

while ( (y not in pos.keys()) and (y not in neg.keys()) ):
    print("Word data not available")
    y = input("Enter the word y: ").lower()
    
z = input("Enter the word z: ").lower()

while ( (z not in pos.keys()) and (z not in neg.keys()) ):
    print("Word data not available")
    z = input("Enter the word z: ").lower()
```

(i) the probability of y being present in the positive sentences

```python
print("The probability of y being present in the positive sentences")
print(f"Probability is { pos[y]/ (pos[y] + neg[y])}")
```

(ii) the joint probability of y and z being present in positive sentences assuming that the presence of y and z in any sentence are independent events.

```python
print("the joint probability of y and z being present in positive sentences assuming that the presence of y and z in any sentence are independent events")
print(f"Probability is { (pos[y]/ (pos[y] + neg[y])) * (pos[z]/ (pos[z] + neg[z]))}")
```
